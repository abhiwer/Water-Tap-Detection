```
def detect(path):
  import cv2
  import matplotlib.pyplot as plt
  %matplotlib inline

  image = cv2.imread(path)
  height, width = image.shape[:2]
  resized_image = cv2.resize(image,(3*width, 3*height), interpolation = cv2.INTER_CUBIC)

  fig = plt.gcf()
  fig.set_size_inches(18, 10)
  plt.axis("off")
  plt.imshow(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB))
  plt.show()
```

```
!./darknet detector test "/content/gdrive/MyDrive/darknet/obj.data" "/content/gdrive/MyDrive/darknet/yolov3-tiny.cfg" "/content/gdrive/MyDrive/darknet/backup/yolov3-tiny_2000.weights" "/content/darknet/depositphotos_20582867-stock-photo-water-tap-with-running-water.jpg" -dont_show

detect('predictions.jpg')
```
